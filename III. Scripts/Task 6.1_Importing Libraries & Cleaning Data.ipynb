### Check for Data Types & change data types

```python
## Check DF infor

df_hotel.info()
```

```python
## Channge Date and Time Columns columns as datetime objects

# Convert Date of Booking column to datetime
df_hotel['Date of Booking'] = pd.to_datetime(df_hotel['Date of Booking'])

# Convert Time column to datetime
df_hotel['Time'] = pd.to_datetime(df_hotel['Time'])

# Convert Check-in date column to datetime
df_hotel['Check-in date'] = pd.to_datetime(df_hotel['Check-in date'])

# Convert Check-Out Date column to datetime
df_hotel['Check-Out Date'] = pd.to_datetime(df_hotel['Check-Out Date'])

# Verify the changes
print(df_hotel.dtypes)
```

```python
df_hotel.head()
```

### Check for Duplicates & remove duplicates

```python
## Check Duplicate values with Booking ID column

df_hotel_duplicates = df_hotel[df_hotel.duplicated('Booking ID',keep=False)]
```

```python
print(df_hotel_duplicates)
```

```python
## Remove duplicates

df_hotel_no_duplicates = df_hotel.drop_duplicates('Booking ID', keep = 'first')
```

```python
df_hotel_no_duplicates.shape
```

### Rename columns

```python
#Renaming collumns

df_hotel_no_duplicates.rename(columns={'Booking Price[SGD]': 'Price per day_Singapore_$', 'GST': 'GST_tax', 'Rooms': 'No_of_rooms_booked', 'Origin Country':'customer_origin_country', 'No of Days': 'Number_of_days_booked', 'Location': 'City'}, inplace=True)
```

```python
df_hotel_no_duplicates.columns
```

### Drop not-relevant columns

```python
## These columns are not relevant to my analysis

Drop_collumns = ['Payment Mode','Bank Name','Check-Out Date', 'Discount', 'GST_tax']
```

```python
df_hotel_no_duplicates.drop(columns = Drop_collumns, inplace=True)
```

```python
df_hotel_no_duplicates.columns
```

### Check for missing Values

```python
missing_values = df_hotel_no_duplicates.isnull().sum()
print(missing_values)
```

### Check for Mixed data types

```python
# Checking for mixed data types

for col in df_hotel_no_duplicates.columns.tolist():
      weird = (df_hotel_no_duplicates[[col]].map(type) != df_hotel_no_duplicates[[col]].iloc[0].apply(type)).any(axis = 1)
      if len (df_hotel_no_duplicates[weird]) > 0:
        print (col)
```

##### No Mixed data types

### Create a new column for 'Room_price_per_day'

```python
df_hotel_no_duplicates['Room_price_per_day'] = df_hotel_no_duplicates['Price per day_Singapore_$'] / df_hotel_no_duplicates['No_of_rooms_booked']
```
