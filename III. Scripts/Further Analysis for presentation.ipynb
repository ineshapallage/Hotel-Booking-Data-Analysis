```python
import pandas as pd

# Assuming your dataset is named df_hotel_cleaned

# Define the age groups and corresponding labels
age_groups = [(19, 30), (31, 50), (51, 70), (71, float('inf'))]
age_group_labels = ['19-30', '31-50', '51-70', '70+']

# Define a function to assign age groups based on age
def assign_age_group(age):
    for i, (lower, upper) in enumerate(age_groups):
        if lower <= age <= upper:
            return age_group_labels[i]
    return 'Unknown'  # Handle any edge cases or missing values

# Apply the function to create the 'Age Group' column
df_hotel_cleaned['Age Group'] = df_hotel_cleaned['Age'].apply(assign_age_group)

# Display the updated DataFrame
print(df_hotel_cleaned.head())
```

```python
import pandas as pd

# Assuming your dataset is named df_hotel_cleaned

# Define the spending categories based on the quartiles
spending_bins = [0, 125, 199, 293, float('inf')]
spending_labels = ['Low', 'Mid-Low', 'Mid-High', 'High']

# Define a function to assign spending categories based on total booking price
def assign_spending_category(price):
    for i, bin_upper in enumerate(spending_bins[1:]):
        if price <= bin_upper:
            return spending_labels[i]
    return 'Unknown'  # Handle any edge cases or missing values

# Apply the function to create the 'Spending Category' column
df_hotel_cleaned['Spending Category'] = df_hotel_cleaned['Total_booking_Price_SG$'].apply(assign_spending_category)

# Display the updated DataFrame
print(df_hotel_cleaned.head())
```

```python
import pandas as pd

# Assuming your dataset is named df_hotel_cleaned

# Define the rating categories based on the quartiles
rating_bins = [0, 4.2, 4.3, 4.5, float('inf')]
rating_labels = ['Low', 'Mid-Low', 'Mid-High', 'High']

# Define a function to assign rating categories based on customer rating
def assign_rating_category(rating):
    for i, bin_upper in enumerate(rating_bins[1:]):
        if rating <= bin_upper:
            return rating_labels[i]
    return 'Unknown'  # Handle any edge cases or missing values

# Apply the function to create the 'Rating Category' column
df_hotel_cleaned['Rating Category'] = df_hotel_cleaned['Hotel Rating'].apply(assign_rating_category)

# Display the updated DataFrame
print(df_hotel_cleaned.head())
```
