```python
### Labeling Correlation heat map 

f = plt.figure(figsize=(8, 8)) # figure size 
plt.matshow(df_hotel_cleaned[Correlation_columns].corr(), fignum=f.number) # type of plot
plt.xticks(range(df_hotel_cleaned[Correlation_columns].corr().shape[1]), df_hotel_cleaned[Correlation_columns].corr().columns, fontsize=12, rotation=90) # x axis labels
plt.yticks(range(df_hotel_cleaned[Correlation_columns].corr().shape[1]), df_hotel_cleaned[Correlation_columns].corr().columns, fontsize=12) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=14) # add font size
plt.title('Correlation Matrix', fontsize=14,) # add title
```

### Q3 Creating matrix heatmap using seaborn

```python

# Create a subplot
fig, ax = plt.subplots(figsize=(10, 8))

# Define the columns to include in the correlation matrix
Correlation_columns = ['Age', 'No. Of People', 'Number_of_days_booked', 'No_of_rooms_booked', 'Hotel Rating', 'Total_booking_Price_SG$', 'Room_price_per_day_SG$', 'Profit Margin']

# Create the correlation matrix heatmap
corr = sns.heatmap(df_hotel_cleaned[Correlation_columns].corr(), annot=True, ax=ax)

# Show the plot
plt.show()
```

#### The correlation matrix shows that,

1.	There is a positive correlation (0.510179) between "Hotel Rating" and "Total_booking_Price_SG$", signaling that hotels with higher ratings tend to have higher booking prices.

2.	There is a negative correlation (-0.484712) between "Hotel Rating" and "Profit Margin", suggesting that hotels with higher ratings may have lower profit margins.

3.	There is a strong negative correlation (-0.626622) between "No_of_rooms_booked" and "Room_price_per_day_SG$", indicating that hotels with more rooms booked tend to have lower room prices per day.

4.	There is a strong negative correlation (-0.619511) between "No_of_rooms_booked" and "Profit Margin", implying that hotels with more rooms booked may have lower profit margins.

5.	There is a strong positive correlation between the number of people and the number of rooms booked. This is because larger groups of people typically require more rooms for accommodation.

### Q7:Create a categorical plot

```python
## Setting up price category column and flags

df_hotel_cleaned.loc[df_hotel_cleaned['Room_price_per_day_SG$'] < 150, 'Price Category'] = 'Low Price'
df_hotel_cleaned.loc[(df_hotel_cleaned['Room_price_per_day_SG$'] >= 150) & (df_hotel_cleaned['Room_price_per_day_SG$'] <= 350), 'Price Category'] = 'Mid Price'
df_hotel_cleaned.loc[df_hotel_cleaned['Room_price_per_day_SG$'] > 350, 'Price Category'] = 'High Price'
```

```python
df_hotel_cleaned.columns
```

```python
## Create a categorical plot

sns.set(style="ticks")
g = sns.catplot(x="Hotel Rating", y="Room_price_per_day_SG$", hue="Price Category", data=df_hotel_cleaned)

plt.savefig('categorical price rating.png')
```

#### The graph shows that when hotel rating goes over 4.2, prices of room per day is increasing. Below 4.2 hotel rating level there are no high price rooms booking.
